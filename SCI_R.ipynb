```r
library(tidyverse)
library(stringr)
```

```r
df <- read_csv("D:/p/sci/sci13.csv")
df<-
df %>%
  select(2,4) %>%
  filter(str_detect(X4, "\\d+")&as.numeric(X4)>0) 
names(df) <- c("journal","IF13")
df$IF13 <- as.numeric(df$IF13)
```

```r
df1 <- read_csv("D:/p/sci/sci14.csv")
df1<-
df1 %>%
  select(2,4) %>%
  filter(str_detect(X4, "\\d+")&as.numeric(X4)>0) 
names(df1) <- c("journal","IF14")
df1$IF14 <- as.numeric(df1$IF14)
```

```r
df2 <- read_csv("D:/p/sci/sci15.csv")
df2<-
df2 %>%
  select(2,4) %>%
  filter(str_detect(X4, "\\d+")&as.numeric(X4)>0) 
names(df2) <- c("journal","IF15")
df2$IF15 <- as.numeric(df2$IF15)
```

```r
df3 <- read_csv("D:/p/sci/sci16.csv")
df3<-
df3 %>%
  select(2,4) %>%
  filter(str_detect(X4, "\\d+")&as.numeric(X4)>0) 
names(df3) <- c("journal","IF16")
df3$IF16 <- as.numeric(df3$IF16)
```

```r
sci<-
df3 %>%
  inner_join(df2,by="journal") %>%
  inner_join(df1,by="journal") %>%
  inner_join(df,by="journal")
sci$journal<- str_trim(sci$journal)
sci <- distinct(sci)
sci$journal<-tolower(sci$journal)
```

```r
sci %>%
  arrange(desc((IF16-IF15)/IF15)) %>%
  filter(IF16>=2) %>%
  filter(str_detect(journal,"(card)|(hyper)|(heart)"))
```
